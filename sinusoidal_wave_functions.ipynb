```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x-axis values
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)

# Base functions
y_base = np.sin(x)
y1 = np.sin(2 * x)                           # frequency doubled
y2 = np.sin(2 * x + np.pi / 3)              # phase shifted
y3 = 3 * np.sin(2 * x + np.pi / 3)          # amplitude modified

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y_base, label=r'$y = \sin(x)$', linestyle='--', color='gray')
plt.plot(x, y1, label=r'$y = \sin(2x)$')
plt.plot(x, y2, label=r'$y = \sin(2x + \frac{\pi}{3})$')
plt.plot(x, y3, label=r'$y = 3\sin(2x + \frac{\pi}{3})$')

# Formatting
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.title('Sinusoidal Function Transformations')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```
