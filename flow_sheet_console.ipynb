```python
# flow_console.py
# Build a simple vertical flow sheet from user inputs using f-strings

print("Enter each step of your process (blank line to finish):")

steps = []
while True:
    s = input("> ").strip()
    if not s:
        break
    steps.append(s)

if not steps:
    print("No steps entered.")
    raise SystemExit

# width for neat alignment
inner = max(len(s) for s in steps) + 4

def box(text: str) -> str:
    top = f"┌{'─' * inner}┐"
    mid = f"│{text.center(inner)}│"
    bot = f"└{'─' * inner}┘"
    return f"{top}\n{mid}\n{bot}"

# print the flow sheet
for i, step in enumerate(steps):
    if i > 0:
        print("      │")
        print("      ▼")
    print(box(step))
```
