```python
import matplotlib.pyplot as plt

def plot_hplc_gradient(time_points, acn_percentages):
    """
    Plots the HPLC gradient based on time points and acetonitrile percentages.

    Parameters:
        time_points (list): List of time points (in minutes).
        acn_percentages (list): List of acetonitrile percentages corresponding to time points.
    """
    # Input validation
    # Ensures that the number of time points matches the number of acetonitrile percentages.
    # If they do not match, a ValueError is raised to alert the user of inconsistent inputs.
    if len(time_points) != len(acn_percentages):
        raise ValueError("time_points and acn_percentages must have the same length.")

    # Plotting the gradient
    # Creates a figure and plots the HPLC gradient using time points and % acetonitrile.
    # The plot includes markers (circles) for each point and lines connecting them to visualize the gradient profile.
    plt.figure(figsize=(8, 6))  # Creates a new figure with a specified size (8x6 inches)
    plt.plot(time_points, acn_percentages, marker='o', linestyle='-', label='HPLC Gradient')  # Plots the data with markers and lines

    # Adding labels and title
    plt.title('HPLC Gradient Profile', fontsize=14)  # Sets the title of the plot with font size 14
    plt.xlabel('Time (minutes)', fontsize=12)  # Labels the x-axis with "Time (minutes)" and font size 12
    plt.ylabel('% Acetonitrile', fontsize=12)  # Labels the y-axis with "% Acetonitrile" and font size 12
    plt.grid(True, linestyle='--', alpha=0.7)  # Adds a grid with dashed lines and partial transparency
    plt.legend(fontsize=10)  # Adds a legend to the plot with font size 10

    # Display the plot
    plt.show()  # Displays the plot on the screen

# Example usage
if __name__ == "__main__":
    try:
        # Prompting the user to enter time points for the HPLC gradient
        print("Enter the time points (in minutes) separated by spaces:")
        time_points = list(map(float, input().split()))  # Reads input, splits it into a list, and converts it to floats

        # Prompting the user to enter corresponding acetonitrile percentages
        print("Enter the corresponding % of acetonitrile (same number of values) separated by spaces:")
        acn_percentages = list(map(float, input().split()))  # Reads input, splits it into a list, and converts it to floats

        # Calls the function to plot the HPLC gradient with the provided inputs
        plot_hplc_gradient(time_points, acn_percentages)
    except ValueError as e:
        # Handles errors related to invalid input
        print(f"Invalid input. Please enter numeric values only. Error: {e}")
```
