**inputs :** solute mass; solution weight.

**output :** solution wt%

```python
# wt% calculator
# Example: You have 2 g of solute in 150 g of total solution

# User inputs
solute_weight_g = float(input("Enter the weight of the solute in g: "))
solution_weight_g = float(input("Enter the total weight of the solution (solute + solvent) in g: "))

# Calculations
solution_percent_wt = round((solute_weight_g / solution_weight_g) * 100, 2)
solvent_weight_g = solution_weight_g - solute_weight_g

# Output
print("\n--- Results ---")
print("Solvent weight: ", solvent_weight_g, "g")
print("Total solution weight: ", solution_weight_g, "g")
print("Solution concentration: ", solution_percent_wt, "% w/w")
```

**inputs :** solvent density; solvent volume; solute weight.

**output :** solution wt%

```python
# wt% calculator
# You want a 1% w/w solution using starch

# User inputs
solvent_density_g_ml = float(input("Enter the density of the solvent in g/ml: "))
solvent_volume_ml = float(input("Enter the volume of the solvent in ml: "))
solute_weight_g = float(input("Enter the weight of the solute in g: "))

# Calculations
solvent_weight_g = solvent_volume_ml * solvent_density_g_ml
solution_weight_g = solute_weight_g + solvent_weight_g
solution_percent_wt = round((solute_weight_g / solution_weight_g) * 100, 2)

# Output
print("The solution weight is", solution_weight_g, "grams")
print("The solution percent wt is", solution_percent_wt, "%")


```
