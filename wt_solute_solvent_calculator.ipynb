**inputs :** solute mass; solution weight.

**output :** solution wt%

```python
# wt% calculator
# Example: You have 2 g of solute in 150 g of total solution

# User inputs
solute_weight_g = float(input("Enter the weight of the solute in g: "))
solution_weight_g = float(input("Enter the total weight of the solution (solute + solvent) in g: "))

# Calculations
solution_percent_wt = round((solute_weight_g / solution_weight_g) * 100, 2)
solvent_weight_g = solution_weight_g - solute_weight_g

# Output
print("\n--- Results ---")
print("Solvent weight: ", solvent_weight_g, "g")
print("Total solution weight: ", solution_weight_g, "g")
print("Solution concentration: ", solution_percent_wt, "% w/w")
```

**inputs :** solution weight; solute weight.

**output :** solution wt%

```python
def calculate_weight_percent():
    try:
        # Get input from user
        solution_weight = float(input("Enter the weight of the solution (in grams): "))
        solute_weight = float(input("Enter the theoretical weight of the solute (in grams): "))

        # Validate inputs
        if solution_weight <= 0:
            print("Solution weight must be greater than 0.")
            return
        if solute_weight < 0:
            print("Solute weight cannot be negative.")
            return

        # Calculate wt%
        wt_percent = (solute_weight / solution_weight) * 100

        # Display result
        print(f"\nWeight percent (wt%) of the solute in the solution: {wt_percent:.2f}%")

    except ValueError:
        print("Please enter valid numeric values.")

# Run the function
calculate_weight_percent()
```

**inputs :** solute mass; solvent density; solvent volume

**output :** solution wt%

```python
def calculate_wt_percent_from_solute_and_solvent():
    try:
        # User inputs
        solute_mass = float(input("Enter the mass of the solute (in grams): "))
        solvent_density = float(input("Enter the density of the pure solvent (in g/mL): "))
        solvent_volume = float(input("Enter the volume of the solvent (in mL): "))

        # Input validation
        if solute_mass < 0 or solvent_density <= 0 or solvent_volume <= 0:
            print("Please enter positive values. Density and volume must be greater than 0.")
            return

        # Calculate mass of solvent
        solvent_mass = solvent_density * solvent_volume

        # Total solution mass
        total_mass = solute_mass + solvent_mass

        # Calculate wt%
        wt_percent = (solute_mass / total_mass) * 100

        # Display result
        print(f"\nWeight percent (wt%) of the solute in the solution: {wt_percent:.2f}%")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

# Run the function
calculate_wt_percent_from_solute_and_solvent()
```
