```python
# inputs
starting_material_MW = float(input("Enter the molecular weight of the starting material (g/mol): "))
starting_material_weight_g = float(input("Enter the weight of the starting material (g): "))

catalyst_MW = float(input("Enter the molecular weight of the catalyst (g/mol): "))
catalyst_equivalents = float(input("Enter the equivalents of catalyst relative to starting material (eq): "))

# calculations
starting_material_moles = starting_material_weight_g / starting_material_MW
catalyst_moles = catalyst_equivalents * starting_material_moles
catalyst_weight_g = catalyst_moles * catalyst_MW
catalyst_moles_percent = catalyst_equivalents * 100

# outputs
print("\n"f"Weight of catalyst to weigh: {catalyst_weight_g:.2f} g")
print(f"Which corresponds to {catalyst_moles:.4f} moles of catalyst")
print(f"Which is {catalyst_equivalents:.2f} equivalents relative to starting material")
print(f"Starting material amount: {starting_material_moles:.4f} moles")
print(f"Catalyst loading: {catalyst_moles_percent:.2f} mol %")
```
