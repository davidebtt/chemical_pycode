```python
# reactor_headspace_calc_pretty.py
# Reactor loading calculator with formatted console output

def box(title, lines, width=50):
    print("┌" + "─" * width + "┐")
    print("│" + title.center(width) + "│")
    print("├" + "─" * width + "┤")
    for line in lines:
        print("│" + line.ljust(width) + "│")
    print("└" + "─" * width + "┘")


print("\nREACTOR HEADSPACE / MAX LOAD CALCULATOR\n")

# ---- INPUT SECTION ----
reactor_volume = float(input("Total reactor volume (L): ").strip())
organic_vw = float(input("Total organic phase (v/w, L per kg SM): ").strip())
water_vw = float(input("Total aqueous phase (v/w, L per kg SM): ").strip())

# ---- CALCULATIONS ----
max_working_volume = reactor_volume * 0.80
headspace = reactor_volume - max_working_volume
total_vw = organic_vw + water_vw
max_sm = max_working_volume / total_vw

# ---- OUTPUT ----
box(
    "INPUT PARAMETERS",
    [
        f"Reactor volume: {reactor_volume:.2f} L",
        f"Organic phase: {organic_vw:.2f} v/w",
        f"Aqueous phase: {water_vw:.2f} v/w",
        f"Total workup volume: {total_vw:.2f} L/kg",
    ],
)

box(
    "REACTOR LIMITS",
    [
        f"Max working volume (80%): {max_working_volume:.2f} L",
        f"Required headspace (20%): {headspace:.2f} L",
    ],
)

box(
    "MAXIMUM STARTING MATERIAL",
    [
        f"Maximum SM load: {max_sm:.2f} kg",
        "",
        "This load ensures total workup volume",
        "remains within 80% reactor capacity.",
    ],
)
```
