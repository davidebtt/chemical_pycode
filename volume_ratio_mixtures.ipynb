This code gives as an output the normalized ratio on 10 units of two solvents miscible in water

**Input 1:** enter the volume/mass ratio of the organic solvent that is miscible with water

**Input 2:** enter the volume/mass ratio of the water of the mixture

```python
def normalize_to_sum_ten(value1, value2):
    total = value1 + value2
    scale = 10 / total
    norm1 = value1 * scale
    norm2 = value2 * scale
    print(f"\nNormalized ratio (organic solvent : water, sum=10): {norm1:.2f} : {norm2:.2f}")
    return norm1, norm2

# Prompt user for input
try:
    organic_solvent_ml = float(input("Enter volume/mass ratio of organic solvent (v/w): "))
    water_ml = float(input("Enter volume of water (v/w): "))

    normalize_to_sum_ten(organic_solvent_ml, water_ml)
except ValueError:
    print("Invalid input. Please enter numeric values for volumes.")
```
