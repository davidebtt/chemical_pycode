```python
# Predefined molecular weights (g/mol) for each reagent
reagents = {
    "H2O": {"molecular_weight": 18.015},
    "NaCl": {"molecular_weight": 58.44},
    "C6H12O6": {"molecular_weight": 180.16}
}

# Prompt user to enter grams used for each reagent
for name, props in reagents.items():
    grams = float(input(f"Enter grams of {name} used: "))
    props["grams_used"] = grams
    props["moles"] = grams / props["molecular_weight"]

# Print result
for name, props in reagents.items():
    print(f"{name}:")
    print(f"  Molecular Weight: {props['molecular_weight']} g/mol")
    print(f"  Grams Used: {props['grams_used']} g")
    print(f"  Moles: {props['moles']:.4f}\n")
```
