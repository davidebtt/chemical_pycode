Use this template dictionary for every compound in your research, once you have the full characterization. A Python dictionary is a very compact way to store a large amount of data in a small block of space. That's how Python works — it allows you to organize and store a lot of information efficiently and clearly.

```python
 compound_1 = {
    "SMILES": "xxx",
    "Molecular Weight": "xxx",
    "Max UV Absorption": "text",
    "TLC (SiO2, Hexane/Ethyl acetate 8/2)": "Rf = 0.xx",
    "1H NMR": "xx",
    "13C NMR": "xx",
    "IR Signals": "xx",
    "Melting Point (°C)": "xx"
}

print(compound_1)
```

```python
compound_1 = {
    "SMILES": input("Enter SMILES: "),
    "Molecular Weight": input("Enter Molecular Weight: "),
    "Max UV Absorption": input("Enter Max UV Absorption: "),
    "TLC (SiO2, Hexane/Ethyl acetate 8/2)": input("Enter TLC value (e.g., Rf = 0.xx): "),
    "1H NMR": input("Enter 1H NMR data: "),
    "13C NMR": input("Enter 13C NMR data: "),
    "IR Signals": input("Enter IR Signals: "),
    "Melting Point (°C)": input("Enter Melting Point (°C): ")
}

print("\nCollected Compound Data:")
print(compound_1)
```
