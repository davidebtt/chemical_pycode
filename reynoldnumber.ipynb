```python
import matplotlib.pyplot as plt
import numpy as np

# Create a log-scale Reynolds number axis
Re = np.logspace(2, 5, 500)  # from 100 to 100,000

# Functions that mimic the qualitative behavior from the diagram
# Normalized curves for visual purposes (not dimensionally exact)

# Power Number (Np): high at low Re, drops in transition, levels off
Np = 1 / (1 + (Re / 4000)**0.8) + 0.15

# Flow Number (Nq): low at low Re, rises sharply, levels off
Nq = (1 / (1 + (4000 / Re)**2)) * 0.8 + 0.2

# Blend Time: high at low Re, decreases rapidly, flat in turbulent
blend_time = 1 / (1 + (Re / 3000)**1.2) + 0.15

# Plotting
plt.figure(figsize=(10, 6))
plt.semilogx(Re, Np, label='$N_p$', color='blue', linewidth=2)
plt.semilogx(Re, Nq, label='$N_q$', color='red', linestyle='--', linewidth=2)
plt.semilogx(Re, blend_time, label='Blend Time', color='green', linestyle=':', linewidth=2)

# Add flow regions
plt.axvspan(100, 4000, color='gray', alpha=0.1, label='Viscous Flow')
plt.axvspan(4000, 10000, color='orange', alpha=0.1, label='Transition Region')
plt.axvspan(10000, 100000, color='cyan', alpha=0.1, label='Turbulent Flow')

# Labels and legend
plt.xlabel("Reynold's Number (Re)")
plt.ylabel("Normalized Value")
plt.title("Dependence of $N_p$, $N_q$, and Blend Time on $Re$")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
