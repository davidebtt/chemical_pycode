```python
def normalize_to_sum_ten(value1, value2):
    total = value1 + value2
    scale = 10 / total
    norm1 = value1 * scale
    norm2 = value2 * scale
    print(f"\nNormalized ratio (acetone : water, sum=10): {norm1:.2f} : {norm2:.2f}")
    return norm1, norm2

# Prompt user for input
try:
    acetone_ml = float(input("Enter volume of acetone (mL): "))
    water_ml = float(input("Enter volume of water (mL): "))

    normalize_to_sum_ten(acetone_ml, water_ml)
except ValueError:
    print("Invalid input. Please enter numeric values for volumes.")
```
