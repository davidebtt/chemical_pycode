Apply the Bragg Law as examples

```python
import numpy as np

# Funzione per calcolare d-spacing da una lista di angoli θ (in gradi)
def calcola_d(angoli_theta_deg, wavelength=1.5406, n=1):
    d_spacings = []
    for theta_deg in angoli_theta_deg:
        theta_rad = np.radians(theta_deg)
        if np.sin(theta_rad) != 0:
            d = n * wavelength / (2 * np.sin(theta_rad))
            d_spacings.append(d)
        else:
            d_spacings.append(None)  # Per gestire θ=0, evitare divisione per zero
    return d_spacings

# Esempio di utilizzo
angoli = [10, 15, 20, 25, 30]  # lista di θ in gradi
d_values = calcola_d(angoli)

for theta, d in zip(angoli, d_values):
    print(f"θ = {theta}°  -->  d = {d:.3f} Å")
```
