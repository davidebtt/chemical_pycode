```python
# flow_console.py
# Build a simple vertical flow from user inputs using f-strings

print("Enter each step of your process (blank line to finish):")
steps = []
while True:
    s = input("> ").strip()
    if not s:
        break
    steps.append(s)

if not steps:
    print("No steps entered.")
    raise SystemExit

# width for neat alignment
inner = max(len(s) for s in steps)

def box(text: str) -> str:
    return f"[{text.center(inner)}]"

# print the flow
for i, step in enumerate(steps):
    if i > 0:
        print("   |")
    print(f" {box(step)}")

```
