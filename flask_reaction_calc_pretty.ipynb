```python
# flask_reaction_calc_pretty.py
# Flask reaction loading calculator (50% max working volume)

def box(title, lines, width=50):
    print("┌" + "─" * width + "┐")
    print("│" + title.center(width) + "│")
    print("├" + "─" * width + "┤")
    for line in lines:
        print("│" + line.ljust(width) + "│")
    print("└" + "─" * width + "┘")


print("\nFLASK REACTION / MAX LOAD CALCULATOR (50% FILL)\n")

# ---- INPUT SECTION ----
flask_volume = float(input("Total flask volume (L): ").strip())
organic_vw = float(input("Total organic phase (v/w, L per kg SM): ").strip())
water_vw = float(input("Total aqueous phase (v/w, L per kg SM): ").strip())

# ---- CALCULATIONS ----
max_working_volume = flask_volume * 0.50   # 50% for flasks
headspace = flask_volume - max_working_volume
total_vw = organic_vw + water_vw

if total_vw <= 0:
    print("Error: total v/w must be greater than zero.")
    raise SystemExit

max_sm = max_working_volume / total_vw

# ---- OUTPUT ----
box(
    "INPUT PARAMETERS",
    [
        f"Flask volume: {flask_volume:.2f} L",
        f"Organic phase: {organic_vw:.2f} v/w",
        f"Aqueous phase: {water_vw:.2f} v/w",
        f"Total workup volume: {total_vw:.2f} L/kg",
    ],
)

box(
    "FLASK LIMITS",
    [
        f"Max working volume (50%): {max_working_volume:.2f} L",
        f"Required headspace (50%): {headspace:.2f} L",
    ],
)

box(
    "MAXIMUM STARTING MATERIAL",
    [
        f"Maximum SM load: {max_sm:.2f} kg",
        "",
        "This load ensures safe mixing,",
        "stirring and avoids overflow.",
    ],
)
```
