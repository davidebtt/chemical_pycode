Description:

This code allows you to solve a proportion of the form A:B = C:x, where

A, B, and  C are known values, and you need to find the value of x.

How to Use:

Run the script.

When prompted, enter the value for A.

Enter the value for B.

Enter the value for C.

The script will calculate and display the value of x (D), rounded to two decimal places.

This is useful for solving basic ratios and proportions in various mathematical problems.

```python
print("A:B = C:x")
A = float(input("Enter number A: "))
B = float(input("Enter number B: "))
C = float(input("Enter number C: "))
x = round((B * C) / A, 2)

print(f"x is {x}")

```
