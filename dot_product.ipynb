```python
from numpy import array, dot, linalg

a = array([1,3])
b = array([4,2])
c = array([2,3])

print("here the vectors")
print(a, b, c)

print("here the dot product of the vector a * b")
d = a.dot(b)
print(d)

print("here the dot product of the vector a * c")
e = a.dot(c)
print(e)

from numpy.linalg import norm  # <-- you can put this on top

print("here the L2 norm of vector a")
l2a = norm(a)
print(l2a)

print("here the L2 norm of vector b")
l2b = norm(b)
print(l2b)

print("here the L3 norm of vector c")
l3c = norm(c, ord=3)
print(l3c)

print("now let´s calculate the angle between vector a and vector b ")

theta = dot(a, b) / (norm(a) * norm(b))
print(theta)

print("now let´s calculate the angle between vector a and vector c ")

gamma = dot(a, c) / (norm(a) * norm(c))
print(gamma)

print("now let´s calculate the angle between vector b and vector c")

delta = dot(b, c) / (norm(b) * norm(c))
print(delta)
```
