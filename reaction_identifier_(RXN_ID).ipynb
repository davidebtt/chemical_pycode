```python
from datetime import datetime

def get_date_input(prompt):
    """ Get a valid date input from the user in the format YYYY-MM-DD. """
    while True:
        date_str = input(prompt)
        try:
            return datetime.strptime(date_str, "%Y-%m-%d").date()
        except ValueError:
            print("Invalid date format. Please enter the date in YYYY-MM-DD format.")

def generate_experiment_id(name, project_type):
    """ Generate a unique RXN-ID using the experiment name, project type, and timestamp. """
    timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
    return f"{name}_{project_type}_{timestamp}"

def main():
    print("### Reaction Identifier (RXN-ID) ###")

    project_type = input("Insert the project type (internal/custom): ").strip().lower()
    name = input("Insert the name of the experiment: ").strip()

    starting_date = get_date_input("Insert the start date (YYYY-MM-DD): ")
    end_date = get_date_input("Insert the end date (YYYY-MM-DD): ")

    experiment_purpose = input("Insert the purpose of your experiment: ").strip()
    results = input("Insert the results of your experiment: ").strip()

    experiment_id = generate_experiment_id(name, project_type)

    print("\n--- Experiment Details ---")
    print(f"Experiment ID: {experiment_id}")
    print(f"Project Type: {project_type.capitalize()}")
    print(f"Name: {name}")
    print(f"Start Date: {starting_date}")
    print(f"End Date: {end_date}")
    print(f"Purpose: {experiment_purpose}")
    print(f"Results: {results}")

if __name__ == "__main__":
    main()


```
