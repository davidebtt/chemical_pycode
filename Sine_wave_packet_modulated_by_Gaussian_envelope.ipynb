```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the wave packet
x0 = 0          # center position
sigma = 1       # width of the Gaussian envelope
k = 5           # wave number (related to momentum)
A = 1           # amplitude

# Spatial coordinates
x = np.linspace(-10, 10, 1000)

# Complex wave packet: Gaussian envelope times complex exponential
psi = A * np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k * x)

# Magnitude and real part
magnitude = np.abs(psi)
real_part = np.real(psi)

# Plotting the wave packet
plt.figure(figsize=(10, 6))
plt.plot(x, real_part, label='Real part (modulated sine wave)')
plt.plot(x, magnitude, 'r--', label='Envelope (magnitude)')
plt.plot(x, -magnitude, 'r--')  # Lower envelope for symmetry
plt.title("Wave Packet: Sine Wave Modulated by Gaussian Envelope")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
