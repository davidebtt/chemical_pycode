```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the wave packet
x0 = 0          # center position
sigma = 1       # width of the Gaussian envelope
k = 5           # wave number (related to momentum)
A = 1           # amplitude

# Spatial coordinates
x = np.linspace(-10, 10, 1000)

# Complex wave packet: Gaussian envelope times complex exponential
psi = A * np.exp(-(x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k * x)

# Magnitude and real part
magnitude = np.abs(psi)
real_part = np.real(psi)

# Plotting the wave packet
plt.figure(figsize=(10, 6))
plt.plot(x, real_part, label='Real part (modulated sine wave)')
plt.plot(x, magnitude, 'r--', label='Envelope (magnitude)')
plt.plot(x, -magnitude, 'r--')  # Lower envelope for symmetry
plt.title("Wave Packet: Sine Wave Modulated by Gaussian Envelope")
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define spherical coordinate grid
phi = np.linspace(0, 2 * np.pi, 200)
theta = np.linspace(0, np.pi, 200)
phi, theta = np.meshgrid(phi, theta)

# s-orbital (l=0, m=0): spherical symmetry
r_s = np.abs(np.sin(theta) * 0 + 1)  # constant amplitude

# p-orbital (l=1, m=0): dumbbell shape along z
r_pz = np.abs(np.cos(theta))  # absolute value for visualization

# Convert to Cartesian coordinates for plotting
def spherical_to_cartesian(r, theta, phi):
    x = r * np.sin(theta) * np.cos(phi)
    y = r * np.sin(theta) * np.sin(phi)
    z = r * np.cos(theta)
    return x, y, z

x_s, y_s, z_s = spherical_to_cartesian(r_s, theta, phi)
x_p, y_p, z_p = spherical_to_cartesian(r_pz, theta, phi)

# Plotting
fig = plt.figure(figsize=(12, 6))

# s orbital
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(x_s, y_s, z_s, rstride=5, cstride=5, color='skyblue', edgecolor='k', linewidth=0.2, alpha=0.8)
ax1.set_title("s Orbital (l=0)")
ax1.set_box_aspect([1,1,1])
ax1.axis('off')

# p_z orbital
ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot_surface(x_p, y_p, z_p, rstride=5, cstride=5, color='orange', edgecolor='k', linewidth=0.2, alpha=0.8)
ax2.set_title("p Orbital (l=1, m=0)")
ax2.set_box_aspect([1,1,1])
ax2.axis('off')

plt.tight_layout()
plt.show()
```
